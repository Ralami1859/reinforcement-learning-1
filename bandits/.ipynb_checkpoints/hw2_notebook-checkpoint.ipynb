## Bernoulli bandit models

Let's define two bandit models with two different complexities:

```python
# Random state
random_state = np.random.randint(1, 312414)

# Different arms of our first MAB
arm11 = arms.ArmBernoulli(0.50, random_state=random_state)
arm12 = arms.ArmBernoulli(0.45, random_state=random_state)
arm13 = arms.ArmBernoulli(0.40, random_state=random_state)
arm14 = arms.ArmBernoulli(0.35, random_state=random_state)

# Different arms of our second MAB
arm21 = arms.ArmBernoulli(0.95, random_state=random_state)
arm22 = arms.ArmBernoulli(0.30, random_state=random_state)
arm23 = arms.ArmBernoulli(0.20, random_state=random_state)
arm24 = arms.ArmBernoulli(0.10, random_state=random_state)

# List of arms (AbstractArm objects)
MAB1 = [arm11, arm12, arm13, arm14]
MAB2 = [arm21, arm22, arm23, arm24]
```

```python
# Compute the complexity of the problem
C1 = compute_complexity(MAB1)
C2 = compute_complexity(MAB2)
print("The complexity of this first MAB is: {}".format(round(C1,2)))
print("The complexity of this second MAB is: {}".format(round(C2,2)))
```

```python
def compute_mean_regrets_algs(MAB, nb_simu=50, T=6000, algs = ["naive", "UCB1", "TS"]):
    """ Perform simulations and compute the mean regret across the simulations
    at each iteration t.
    """
    # Store the mean regrets at each iteration t for each algorithm
    mean_regrets_algs = {}
    
    for alg in algs:  
        mean_regrets_algs[alg] = compute_mean_regrets(MAB=MAB, T=T, nb_simu=nb_simu, algorithm=alg)
            
    return mean_regrets_algs
```

```python
def plot_regrets(MAB, T, nb_simu, mean_regrets_algs, to_plot = ["naive", "UCB1", "TS", "oracle"]):
    """Plot the cumulative regrets resulting from different algorithms on the 
    same MAB (same time horizon and same number of simulations): naive 
    strategy, UCB1 or Thompson Sampling.
    """
    
    # Iterations for each simulation (x coordinate)
    iterations = np.arange(T)

    if "naive" in set(mean_regrets_algs).intersection(to_plot):
        plt.plot(iterations, mean_regrets_algs["naive"].cumsum(), label = "Naive strategy")
    if "UCB1" in set(mean_regrets_algs).intersection(to_plot):
        plt.plot(iterations, mean_regrets_algs["UCB1"].cumsum(), label = "UCB1")
    if "TS" in set(mean_regrets_algs).intersection(to_plot):
        plt.plot(iterations, mean_regrets_algs["TS"].cumsum(), label = "Thompson")

    # Compute the complexity of the problem C(p)
    c = compute_complexity(MAB)
    # Add the oracle regret curve t -> C(p)log(t)
    if "oracle" in to_plot:
        oracle = [c*np.log(t) for t in iterations]
        plt.plot(iterations, oracle, label = "Oracle regret curve")
        
    plt.xlabel("Rounds")
    plt.ylabel("Mean Cumulative Regret")
    plt.title("Estimated cumulative regret with {} simulations".format(nb_simu))

    plt.legend()

    plt.show()
```

```python
# Time horizon
T = 10000

# Number of simulations
nb_simu = 100

# Perform simulations and compute the mean regrets for the three algorithms:
# the naive strategy, UCB1 and Thompson Sampling
print("--> Performing simulations for the first problem (C={})".format(round(C1)))
mean_regrets_algs_1 = compute_mean_regrets_algs(MAB1, nb_simu=nb_simu, T=T)
print("--> Performing simulations for the second problem (C={})".format(round(C2)))
mean_regrets_algs_2 = compute_mean_regrets_algs(MAB2, nb_simu=nb_simu, T=T)
```

```python
# Plot the expected regret for the first MAB
print("--> Plotting the expected regret for the first problem (C={})".format(round(C1)))
plot_regrets(MAB1, T=T, nb_simu=nb_simu, mean_regrets_algs=mean_regrets_algs_1)
```

```python
# Plot the expected regret for the second MAB
print("--> Plotting the expected regret for the second problem (C={})".format(round(C2)))
plot_regrets(MAB2, T=T, nb_simu=nb_simu, mean_regrets_algs=mean_regrets_algs_2)
```

> The above plots show that the final regrets for both *UCB1* and *Thompson Sampling* algorithms are lower in the "easiest" problem (C=2).

> Also, we can notice that Lai and Robbins' is in fact a lower bound on the *UCB1* mean cumulative regret after a certain round ( $T \rightarrow \infty$). For the first problem, this occurs after 4000 rounds, for the second one, its occurs very early. However, the *Thompson Sampling* mean cumulative regret is always under the bound. This can be explained by the fact that T is not big enough (a very big number of rournds may result in the intersection of the red and green curves).

## Non-parametric bandits 

> In order to deal with non Bernoulli arms, an implementation of *Thompson Sampling* consists in the following change from a "classic" *Thompson Sampling*: the sampling returns directly the sampled reward of the arm if it is a Bernoulli arm. Otherwise, we draw a reward from a Bernoulli distribution with the sampled reward as a parameter. This has been implemented the **sample_reward** method of the class **MABAlgorithm**.

> In this case, the notion of complexity does not really make sense because the function *kl* has been defined for two Bernoulli distributions.

Let's define a bandit model with arms following exponential distributions:

```python
# Random state
random_state = np.random.randint(1, 312414)

# Different arms of our first MAB
arm31 = arms.ArmExp(L=0.5, random_state=random_state)
arm32 = arms.ArmExp(L=1, random_state=random_state)
arm33 = arms.ArmExp(L=1.5, random_state=random_state)
arm34 = arms.ArmExp(L=2, random_state=random_state)

# List of arms (AbstractArm objects)
MAB3 = [arm31, arm32, arm33, arm34]
```

```python
# Compute the complexity of the problem (eventhough we cannot really use the 
# previous formula here)
C3 = compute_complexity(MAB3)
print("The complexity of this first MAB is: {}".format(round(C3,2)))
```

```python
# 100 simulations with T=10000
mean_regrets_algs_3 = compute_mean_regrets_algs(MAB3, nb_simu=nb_simu, T=T)
```

```python
# Plot the expected regret
plot_regrets(MAB2, T=T, nb_simu=nb_simu, mean_regrets_algs=mean_regrets_algs_3)
```

# Linear bandit on real data 

```python
from linMAB import linUCB, random_policy, greedy_policy
from linearmab_models import LinearMABModel, ColdStartMovieLensModel
```

```python
# Random state
random_state = np.random.randint(0, 24532523)
```

```python
# Cold start problem (movie lens)
movie_model = ColdStartMovieLensModel(random_state=random_state, noise=0.1)
```

```python
# Features (nm, d) where nm is the number of movies, d the feature's dimension
# resulting from the matrix factorization
movie_model.features.shape
```

```python
# Real theta dimension for the movie lens problem 
movie_model.real_theta.shape
```

```python
# Cross-validation: LinUCB on the cold start movie problem with different 
# parameters of alpha (that controls the confidence interval)

alphas = [2, 2.5, 3, 3.5, 4]
mean_regrets = {}
mean_norms = {}
for alpha in alphas:
    mean_regrets[alpha], mean_norms[alpha] = linUCB(movie_model, lambda_=1, alpha = alpha)
```

```python
# Plot the two metrics for evaluating the algorithms

plt.figure(1)
plt.subplot(121)
for alpha in alphas:
    plt.plot(mean_norms[alpha], label=r"$\alpha = {}$".format(alpha))
plt.ylabel('d(theta, theta_hat)')
plt.xlabel('Rounds')
plt.legend()

plt.subplot(122)
for alpha in alphas:
    plt.plot(mean_regrets[alpha].cumsum(), label=r"$\alpha = {}$".format(alpha))
plt.ylabel('Cumulative Regret')
plt.xlabel('Rounds')
# Locate legend at the figure’s top right-hand corner instead of the 
# axes’ corner
plt.legend(bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)

plt.show()
```

We will choose a parameter $\alpha = 4$ since the l2-norm of the estimated $\theta$ with respect to the true one converges faster (and is lower).

```python
# Peform 50 simulations (T=6000) for the three different algorithms
mean_regrets_linUCB, mean_norms_linUCB = linUCB(movie_model)
mean_regrets_random, mean_norms_random = random_policy(movie_model)
mean_regrets_greedy, mean_norms_greedy = greedy_policy(movie_model)
```

```python
# Plot the mean l2-norm of the estimated theta w.r.t. the true one
plt.plot(mean_norms_linUCB, label="LinUCB")
plt.plot(mean_norms_random, label="random")
plt.plot(mean_norms_greedy, label="$\epsilon$-greedy")
plt.ylabel("d(theta, theta_hat)")
plt.xlabel("Rounds")
plt.legend()
plt.show()
```

```python
# Plot the expected cumulative regret
plt.plot(mean_regrets_linUCB.cumsum(), label="LinUCB")
plt.plot(mean_norms_random.cumsum(), label="random")
plt.plot(mean_norms_greedy.cumsum(), label="$\epsilon$-greedy")
plt.ylabel("Cumulative Regret")
plt.xlabel("Rounds")
plt.legend()
plt.show()
```

> The above curves show the random strategy converges faster towards the real $\theta$ since there is a lot of exploration (repeatedly pulls all the arms). However, the regret is very high compared to the two other methods because of the lack of exploitation (we pull a lot of suboptimal arms and thus do not recommend good movies for the users).

> The $\epsilon$-greedy has more exploitation (because it pulls optimal arms with probability $1-\epsilon$, so the regret is lower than the random's one. However, the convergence towards the real $\theta$ is lower.

> With $\alpha=4$, the *LinUCB* algorithm does not really converge towards the real $\theta$ (further than the two others), but the expected regrets are lower. We may have to use higher values for $\alpha$ in order to get closer to the real $\theta$.
